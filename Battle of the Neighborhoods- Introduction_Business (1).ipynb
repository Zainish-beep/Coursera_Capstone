## Introduction:

The purpose of this project is to compare the financial capitals of South Korea and Japan, Seoul and Tokyo. Seeing as these two cities are the most populated cities in their respective countries, this project hopes the explore the similarities and dissimilarities of each. This will allow us to gain further understanding of what characteristics overlap for those countries to be as big and important as they are, and the dissimalirities will help us gain insight to the differences in the countries.
