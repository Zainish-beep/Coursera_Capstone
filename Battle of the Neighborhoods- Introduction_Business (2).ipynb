# Business Proposal for Battle of the Neighborhoods- Seol & Toronto

## Introduction:

The purpose of this project is to compare the financial capitals of South Korea and Canada, Seoul and Toronto. Seeing as these two cities are the most populated cities in their respective countries, this project hopes the explore the similarities and dissimilarities of each. This will allow us to gain further understanding of what characteristics overlap for those countries to be as big and important as they are, and the dissimalirities will help us gain insight to the differences in the countries. 

# Who this could be helpful for

This could be helpful information to people who possibly want to move to Canada from South Korea and see if there are similar opportunities and lifestyles. Many Koreans come to study abroad in Canada, especially go to univeristies in Canada, so this project could provide them with that information and give them more insights.  

# Problem: How similar/dissimilar are the neighborhoods

This project will answer the question, and show us how similarly or differently they are clustered, how the venues differ and we can surmise how much adjusting would a Korean have to do if they moved to Toronto.

# Foursquare API

Foursquare will be the main resource used to find location data in Korea and Toronto. It will help find nearby venues and building a map will help visualize the clustered neighborhoods
